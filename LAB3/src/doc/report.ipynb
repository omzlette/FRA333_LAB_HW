## **<font size=6>Algorithm X สำหรับสั่งงานหุ่นยนต์</font>**

การทำ Algorithm X สำหรับสั่งงานหุ่นยนต์ จะเริ่มจากการเขียน Node ชื่อ "SandeStatePub"

![Node](pic/SandeStatePubNode.png)

สร้าง Subscriber มารับค่า IMU ผ่าน ROS Topic ที่ชื่อว่า /sandevistan_IMU 

![Calibrate Node](pic/IMUSub.png)

และนำค่าที่ได้ไปลบกับค่า Mean ของ Linear Acceleration และ Angular Velocity ในแต่ละแกน และเก็บค่าไว้ใน Array "self.linear_accel" และ "self.angular_vel"

![Calibrate Node](pic/Subcallback.png)

จากนั้นสร้าง Array "self.pos" และ "self.oldpos" มาเก็บค่าตำแหน่งเริ่มต้นของ End Effector ใน Cartesian Coordinate

![Calibrate Node](pic/posoldpos.png)

และแปลงค่าตำแหน่งปัจจุบันของ End Effector ใน Array "self.pos" ที่อยู่ใน Cartesian Coordinate ให้กลายเป็น ตำแหน่งใน Cylindrical Coordinate และเก็บไว้ใน Array "self.poscylin"

![Calibrate Node](pic/cartetocylin.png)

ต่อมานำค่า Linear Acceleration ใน Array "self.linear_accel" ในแต่ละแกนมาเปรียบเทียบ เพื่อนำมาบังคับหุ่นยนต์ ใน Cylindrical Coordinate ตามทิศทางที่ต้องการดังนี้
- เมื่อ X เป็นบวก หุ่นยนต์จะยืดไปด้านหน้า
- เมื่อ X เป็นลบ หุ่นยนต์จะหดไปด้านหลัง
- เมื่อ Y เป็นบวก หุ่นยนต์จะหมุนทวนเข็มนาฬิกา
- เมื่อ Y เป็นลบ หุ่นยนต์จะหมุนตามเข็มนาฬิกา
- เมื่อ Z เป็นบวก หุ่นยนต์จะยืดไปด้านบน
- เมื่อ Z เป็นลบ หุ่นยนต์จะหดไปด้านล่าง

![Calibrate Node](pic/newposcylin.png)

หลังจากนั้น จะนำค่าตำแหน่งที่เปลี่ยนแปลงของ End Effector ใน Cylindrical Coordinate มาแปลงกลับให้เป็น Cartesian Coordinate และนำไปเก็บไว้ใน Array "self.pos"

![Calibrate Node](pic/cylintocarte.png)

และนำค่าตำแหน่งที่เปลี่ยนแปลงของ End Effector มาเช็คกับ Workspace ของหุ่นยนต์
หากตำแหน่งอยู่ใน Workspace 
- ทำการคำนวณ Joint Configuration (q1, q2, q3) มาเก็บไว้ใน Array "self.qarray"
- อัพเดตตำแหน่งปัจจุบันของ End Effector ไว้ใน Array "self.oldpos"
หากตำแหน่งอยู่นอก Workspace
- ไม่คำนวณ Joint Configuration ของตำแหน่งที่เปลี่ยนแปลง ทำให้หุ่นยนต์ไม่ขยับ
- เปลี่ยนค่าตำแหน่งที่เปลี่ยนแปลงของ End Effector ใน Array "self.pos" ให้เป็นตำแหน่งปัจจุบันของ End Effector ใน Array "self.oldpos"

![Calibrate Node](pic/calculateq.png)

และสุดท้าย Node จะทำการ Publish ค่า Joint Configuration ไปยัง Gazebo ผ่าน Topic ชื่อ "/joint_trajectory_position_controller/joint_trajectory"
